```python
import pandas as pd
import tensorflow
import numpy as np
```

```python
data=pd.read_csv('/content/sentiment_analysis update.csv')
```

```python
data.head()
```

drop unnsee column

```python
data.drop(columns=['Year','Month','Day','Time of Tweet',"Platform"],inplace=True,axis=1)
```

```python
data.head()
```

```python
data.head()
```

convert into tect using label encoder

```python
import sklearn
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data['sentiment']=le.fit_transform(data['sentiment'])
```

```python
data['sentiment'].value_counts()
```

```python
data.head()
```

positive->2
Neuteal->1
Negative->0

split into trainig and testing

```python
from sklearn.model_selection import train_test_split
train_data,test_data=train_test_split(data,test_size=0.2)
```

```python
test_data.shape
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
tokenizer=Tokenizer()
```

```python
tokenizer.fit_on_texts([train_data["text"].tolist()])
```

```python
x_train=pad_sequences(tokenizer.texts_to_sequences(train_data["text"]),maxlen=50)
x_test=pad_sequences(tokenizer.texts_to_sequences(test_data["text"]),maxlen=50)
```

```python
x_train
```

```python
y_train=train_data['sentiment']
y_test=test_data['sentiment']
```

applt LSTM

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM,Dense,Embedding
```

```python
model=Sequential()
```

```python
model.add(Embedding(input_dim=1000,output_dim=100,input_length=50)),
model.add(LSTM(32)),
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=50,validation_split=0.2,batch_size=32)
```

```python
loss,accuracy=model.evaluate(x_test,y_test)
```

```python
print(f"{loss}")
print(f"{accuracy}")
```

```python
from collections.abc import Sequence
def pre_sen(review):
  sequence=tokenizer.texts_to_sequences(review)
  padded_s=pad_sequences(sequence,maxlen=50)
  prediction=model.predict(padded_s)
  senti='positive' if prediction[0][0]>1 else "negative"
  return senti
```

```python
re='Dont angry me	'
senti=pre_sen(re)
print(f"{senti}")
```
